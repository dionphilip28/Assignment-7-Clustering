### Problem Statement: Perform Clustering(Hierarchical, Kmeans & DBSCAN) for the crime data and identify the number of clusters formed.

## Hierarchical Clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
crime = pd.read_csv('crime_data.csv')
crime
```

```python
crime.rename({'Unnamed: 0':'Place'},axis=1,inplace=True)
```

```python
crime_df = crime.copy()
crime_df
```

```python
crime_df.drop('Place',axis=1,inplace=True)
```

```python
crime_df.head()
```

```python
#Normalization function
def min_max(i):
    x = (i-i.min())/(i.max()-i.min())
    return x
```

```python
#Normalized dataframe (Considering only numerical part of data)
crime_norm = min_max(crime_df.iloc[:,:])
```

```python
crime_norm
```

```python
#Create dendrogram
fig, axes = plt.subplots(figsize=(15, 8))
dendrogram = sch.dendrogram(sch.linkage(crime_norm, method='complete'))
```

```python
#Create clusters
hc = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='complete')
```

```python
hc
```

```python
#Save clusters for charts
y_hc = hc.fit_predict(crime_norm)
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
```

```python
Clusters
```

```python
Clusters.value_counts()
```

```python
hc.labels_
```

```python
crime1 = crime_df.copy()
crime1
```

```python
crime1['h_clusterid']=y_hc
```

```python
crime1
```

```python
crime1.groupby('h_clusterid').agg(['mean']).reset_index()
```

```python
plt.figure(figsize=(5,5))  
plt.scatter(crime1['h_clusterid'],crime1['Assault'], c=y_hc)
plt.show()
```

```python
plt.figure(figsize=(5,5))  
plt.scatter(crime1['h_clusterid'],crime1['UrbanPop'], c=y_hc)
plt.show()
```

```python
sns.pairplot(crime1)
```

## K-Means Clustering

```python
from matplotlib import pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
array = crime_df.values
array
```

```python
#Normalizing the data
scaler = StandardScaler()
X = scaler.fit_transform(array)
```

```python
X
```

```python
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11),wcss,marker='o')
plt.title("Elbow Method")
plt.xlabel('No.of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

for i in range(2,11):
    cluster = KMeans(n_clusters=i,random_state=0)
    cluster.fit_predict(X)

    silhouette_avg = silhouette_score(X, cluster.labels_, metric='euclidean')
    print('For no. of clusters =', i, 
           '\nThe average silhouette score is :', silhouette_avg)
```

```python
#Build cluster algorithm
clusters_new = KMeans(4, random_state=0)
clusters_new.fit(X)
```

```python
clusters_new.labels_
```

```python
clusters_new.cluster_centers_
```

```python
crime2 = crime_df.copy()
crime2
```

```python
crime2['k_clusterid'] = clusters_new.labels_
```

```python
crime2
```

```python
crime2.groupby('k_clusterid').agg(['mean']).reset_index()
```

```python
plt.figure(figsize=(5,5))  
plt.scatter(crime2['k_clusterid'],crime2['UrbanPop'], c=clusters_new.labels_)
plt.show()
```

## DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn import metrics
```

```python
crime3 = crime_df.copy()
crime3
```

```python
crime_df.head()
```

```python
#Normalized data
X
```

```python
dbscan = DBSCAN(eps=1, min_samples=4)
dbscan.fit(X)
```

```python
dbscan.labels_
```

```python
#Silhouette Score
S=metrics.silhouette_score(X,dbscan.labels_)
S
```
